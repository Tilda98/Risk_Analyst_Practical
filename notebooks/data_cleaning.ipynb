```python
# Renaming column names in call center data

call_center.rename(columns={'Rafiki ID': 'Rafiki Id',
                   'What is your gender?': 'Gender', 
                   'What year were you born?': 'Year of birth',
                   'Did you recently visit a Tiko health facility to access a service?': 'Visited facility',
                   'Were you satisfied with the service you took?': 'Satisfaction',
                   'Please share any other feedback you have concerning the service you took, or the facility (clinic/chemist) you visited': 'Feedback',
                   'Do you have any additional questions or concerns about family planning or Tiko?': 'Concerns',
                   'Write any intelligence collected from the call': 'Notes'}, inplace=True)
call_center.head()
```
