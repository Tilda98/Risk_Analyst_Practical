```python
# Checking for unique values in year of birth column
merged_data['Year of birth'].unique()
```

But first, as you see above, it seems that the "Year of birth" column contains non-numeric values in ranges such as '2004-2009'. I will clean this column by converting valid year ranges to a single year or removing invalid entries.

```python
# Replace year ranges with approximate median values
merged_data['Year of birth'].replace({'2004-2009': 2006, '1998-2003': 2000, 'Before 1998': 1997}, inplace=True)

# Convert 'Year of birth' to numeric, forcing errors to NaN
merged_data['Year of birth'] = pd.to_numeric(merged_data['Year of birth'], errors='coerce')

# Fill NaN values with the median of the cleaned numeric values
merged_data['Year of birth'].fillna(merged_data['Year of birth'].median(), inplace=True)
```

I'll proceed with cleaning the data

```python
# Replace missing values for other columns
merged_data['Gender'].fillna('Unknown', inplace=True)
merged_data['Visited facility'].fillna('Unknown', inplace=True)
merged_data['Tiko card'].fillna('Unknown', inplace=True)
merged_data['Satisfaction'].fillna('Not provided', inplace=True)
merged_data['Feedback'].fillna('Not provided', inplace=True)
merged_data['Concerns'].fillna('None reported', inplace=True)
```

```python
# Confirming missing values

merged_data.isnull().sum()
```

```python
# Deleting the rafiki Id and Event Id columns
del merged_data['Rafiki Id'],
del merged_data['Event Id']
merged_data.head()
```

```python
# Converting data types
merged_data['Offer Redemption Date'] = pd.to_datetime(merged_data['Offer Redemption Date'], errors='coerce')

# Calculate age based on 'Year of birth' and 'Offer Redemption Date'
current_year = datetime.datetime.now().year
merged_data['Age'] = current_year - merged_data['Year of birth']

# Removing duplicate entries
merged_data.drop_duplicates(inplace=True)

# Standardizing text data
merged_data['Call status'] = merged_data['Call status'].str.capitalize()
merged_data['Feedback'] = merged_data['Feedback'].str.capitalize()

merged_data.head()
```

```python
# Dropping the year of birth column to remain with the age column
del merged_data['Year of birth']
```
